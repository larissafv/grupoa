```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
tab = pd.read_csv('caso_full.csv')
```

```python
tab.dropna(axis=0,subset=['city_ibge_code'],inplace=True)
tab = tab.astype({'city_ibge_code':int},copy=False)
tab
```

```python
tab = tab.drop(columns=['is_last', 'is_repeated', 'new_confirmed', 'new_deaths', 'estimated_population_2019', 'last_available_confirmed_per_100k_inhabitants', 'last_available_death_rate', 'order_for_place'])
```

```python
ultimo_dia_semana = list(tab[tab['city']=='São Paulo'].groupby('epidemiological_week')['date'].agg(lambda x:x[x.last_valid_index()]).values)
ultimo_dia_semana
```

```python
tab_states = tab[tab['place_type']=='state']
tab_states
```

```python
df_states = tab_states[tab_states['epidemiological_week']==31].groupby('state')[['city_ibge_code']].agg(lambda x:x[x.last_valid_index()]).copy()
df_states
```

```python
for x in range(9,35):
    df_states['last_available_confirmed_week_{}'.format(x)] = tab_states[tab_states['epidemiological_week']==x].groupby('state')['city_ibge_code', 'last_available_confirmed'].agg(lambda x:x[x.last_valid_index()]).last_available_confirmed
    df_states['last_available_deaths_week_{}'.format(x)] = tab_states[tab_states['epidemiological_week']==x].groupby('state')['city_ibge_code', 'last_available_deaths'].agg(lambda x:x[x.last_valid_index()]).last_available_deaths
    
df_states = df_states.fillna(0)
df_states
```

```python
total_confirmed = tab_states[tab_states['date'].isin(ultimo_dia_semana)].groupby('epidemiological_week')['last_available_confirmed'].agg(sum)
total_deaths = tab_states[tab_states['date'].isin(ultimo_dia_semana)].groupby('epidemiological_week')['last_available_deaths'].agg(sum)
total_confirmed
```

```python
dict = {'city_ibge_code': 0}
for x in range (26):
    dict['last_available_confirmed_week_{}'.format(x+9)] = total_confirmed.values[x:x+1]
    dict['last_available_deaths_week_{}'.format(x+9)] = total_deaths.values[x:x+1]
    
df_br = pd.DataFrame(data=dict, index=['TOTAL'])
df_br
```

```python
df_states = pd.concat([df_states, df_br])
df_states
```

```python
df_states.to_csv('casos_e_mortes_acumulados_estado.csv')
```

```python
tab_muni = tab[tab['place_type']=='city']
tab_muni
```

```python
df_muni = tab_muni[tab_muni['epidemiological_week']==31].groupby('city_ibge_code')['city', 'state'].agg(lambda x:x[x.last_valid_index()]).copy()
df_muni
```

```python
for x in range(9,35):
    df_muni['last_available_confirmed_week_{}'.format(x)] = tab_muni[tab_muni['epidemiological_week']==x].groupby('city_ibge_code')['last_available_confirmed'].agg(lambda x:x[x.last_valid_index()])
    df_muni['last_available_deaths_week_{}'.format(x)] = tab_muni[tab_muni['epidemiological_week']==x].groupby('city_ibge_code')['last_available_deaths'].agg(lambda x:x[x.last_valid_index()])
    
df_muni = df_muni.fillna(0)
df_muni
```

```python
df_muni.to_csv('casos_e_mortes_acumulados_municipio.csv')
```
