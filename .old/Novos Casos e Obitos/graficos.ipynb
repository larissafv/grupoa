```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Colocando alguns defaults para embelezar os plots!
sns.set_style("darkgrid")
sns.set_palette('bright')
```

```python
df = pd.read_csv('../data/novos_casos_obitos.csv')

# Os dados da semana 35 parecem não estar completos, pois estão muito discrepantes em todos os estados
df = df[ df['epidemiological_week'] != 35 ]
df
```


# Foram feitas as mesmas analises por regiao retirando alguns estados que estavam com dados discrepantes do restante da região

```python
centro_oeste = ['GO','MT','MS','DF']
sudeste = ['MG','ES','RJ','SP']
sul = ['RS','SC','PR']
nordeste = ['AL','BA','CE','MA','PB','PE','PI','RN','SE']
norte = ['AM','RR','RO','AP','PA','TO','AC']


regioes = {'Sul': sul,
      'Sudeste': sudeste,
      'Centro Oeste': centro_oeste,
      'Norte': norte,
      'Nordeste': nordeste}
```

<h2>Novas Mortes<h2>

```python
######### Novas Mortes #############

for regiao in regioes:
    df_regiao = df[df['state'].isin(regioes[regiao])]
    
    f = sns.factorplot(data= df_regiao, x= 'epidemiological_week', y= 'new_deaths', hue= 'state', legend=False)
    f.fig.set_size_inches(40, 20)
    plt.title("Novas mortes por semana epidemiológica - "+regiao, fontsize= 40)
    f.set_xlabels('Semana Epidemiológica', fontsize= 40)
    f.set_ylabels('Novas Mortes', fontsize= 40)
    plt.legend(title='estado',loc='upper left',fontsize=30, title_fontsize='40', frameon=True)
    plt.savefig('grupoa/new_deaths/new_deaths_'+regiao+'.png')
    plt.show()
```

<h2>Novos casos<h2>

```python
######### Novos Casos #############

for regiao in regioes:
    df_regiao = df[df['state'].isin(regioes[regiao])]
    
    f = sns.factorplot(data= df_regiao, x= 'epidemiological_week', y= 'new_confirmed', hue= 'state', legend=False)
    f.fig.set_size_inches(40, 20)
    plt.title("Novos casos confirmados por semana epidemiológica - "+regiao, fontsize= 40)
    f.set_xlabels('Semana Epidemiológica', fontsize= 40)
    f.set_ylabels('Novos Casos Confirmados', fontsize= 40)
    plt.legend(title='estado',loc='upper left',fontsize=30, title_fontsize='40', frameon=True)
    plt.savefig('grupoa/new_cases/new_cases_'+regiao+'.png')
    plt.show()
```

```python
######### Novos Casos - Retirando pontos fora da curva #############

df_regiao = df[df['state'].isin(nordeste)]

f = sns.factorplot(data= df_regiao[ ~df_regiao['state'].isin(['BA']) ], x= 'epidemiological_week', y= 'new_confirmed', hue= 'state', legend=False)
f.fig.set_size_inches(40, 20)
plt.title("Novos casos confirmados por semana epidemiológica - Nordeste sem BA", fontsize= 40)
f.set_xlabels('Semana Epidemiológica', fontsize= 40)
f.set_ylabels('Novos Casos Confirmados', fontsize= 40)
plt.legend(title='estado',loc='upper left',fontsize=30, title_fontsize='40', frameon=True)
plt.savefig('grupoa/new_cases/filter/new_cases_Nordeste-sem-BA.png')
plt.show()
```
