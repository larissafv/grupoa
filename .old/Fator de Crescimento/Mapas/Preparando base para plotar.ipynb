```python
import geopandas as gpd
import geobr
import pandas as pd
```

```python
municipios = pd.read_csv("Novos_Casos_Por_Cidade.csv")
municipios
```

```python
cod_mun = []
for i in municipios['city_ibge_code'].unique():
    cod_mun.append(i)

semanas = []
for i in municipios['epidemiological_week'].unique():
    semanas.append(i)
```

```python
codigos = []
novos_confirmados = []
semana_epidemiologica = []

for c in cod_mun:
    mun_c = municipios[municipios['city_ibge_code'] == c]
    mun_c = mun_c.groupby(['epidemiological_week']).sum()
    for n in mun_c['new_confirmed'].tolist():
        novos_confirmados.append(n)
    for i in mun_c.index:
        semana_epidemiologica.append(i)
        codigos.append(c)

soma = pd.DataFrame(data = {'city_ibge_code': codigos, 'epidemiological_week': semana_epidemiologica,
                           'new_confirmed': novos_confirmados})
soma
```

```python
codigos = []
semana_epidemiologica = []
fator = []
for c in cod_mun:
    mun_c = soma[soma['city_ibge_code'] == c]
    linhas = mun_c.values
    for i in range(1,len(mun_c)):
        if(linhas[i-1][2] != 0):
            codigos.append(c)
            semana_epidemiologica.append(linhas[i][1])
            fator.append(round((linhas[i][2]/linhas[i-1][2]),2))

fator_crescimento = pd.DataFrame(data = {'city_ibge_code': codigos, 'epidemiological_week': semana_epidemiologica,
                                        'growth_factor': fator})
fator_crescimento
```

```python
fator_crescimento.to_csv("Fator_Crescimento_Municipios.csv", sep = ";", index = False)
```
